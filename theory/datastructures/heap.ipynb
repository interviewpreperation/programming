```python
from heapq import heappush, heappop, heapify
```

# Min Heap

```python
# Example of a min-heap (smallest element at the root)
heap = []

# Insert elements into the heap
heappush(heap, 10)
heappush(heap, 5)
heappush(heap, 15)
heappush(heap, 7)

print(heap)
```

```python
# Get the smallest element (root of the heap)
smallest = heappop(heap)
print("Smallest element:", smallest)

# Get the smallest element without popping it
next_smallest = heap[0]
print("Next smallest element (without popping):", next_smallest)
```

```python
# Convert a list to a heap
data = [13, 2, 7, 14, 6]
heapify(data)
print("Heapified data:", data)
```

```python
# Heapsort: converting a heap to a sorted list
sorted_list = []
while data:
    sorted_list.append(heappop(data))
print("Sorted list using heapsort:", sorted_list)
```

```python
# Example of a max-heap (simulate by storing negative values)
max_heap = []

# Insert elements into the max-heap (store negatives)
heappush(max_heap, -10)
heappush(max_heap, -5)
heappush(max_heap, -15)
heappush(max_heap, -7)

print(max_heap)
```
